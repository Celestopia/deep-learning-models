```python
import numpy as np
import matplotlib.pyplot as plt
import os 


# 指定文件夹路径
download_path = "D:\\SRT_Dataset\\Data_Download"

# 获取文件夹下的所有文件名称
folder_names = os.listdir(download_path) # Tail_652_1,Tail_652_2,...

mat_paths=[]
for folder_name in folder_names:
     folder_path=os.path.join(download_path,folder_name) # mat所在文件夹的存储路径
     mat_names=os.listdir(folder_path)
     for mat_name in mat_names:
         mat_id=mat_names.index(mat_name) # 该mat文件的序号，从0开始
         mat_paths.append(os.path.join(folder_path,mat_name))
print(len(mat_paths))
```

```python
import scipy.io as scio
normal_mat_info=[]

#for i in range(lower_bound,upper_bound):#在指标范围内搜索mat
for i in range(0,2000):#在指标范围内搜索mat
    mat_path=mat_paths[i]
    mat_data=scio.loadmat(mat_paths[i]) # 数据类型：dict
    wind_shear_data=mat_data['WSHR'][0][0][0]
    mat_data_len=wind_shear_data.shape[0]
    score=0 # 用于评估该mat文件是否正常
    for j in range(mat_data_len):
        x1=mat_data['WD'][0][0][0][4*j][0]
        x2=mat_data['WS'][0][0][0][4*j][0]
        x3=mat_data['TAS'][0][0][0][4*j][0]
        x4=mat_data['GS'][0][0][0][4*j][0]
        if x1!=0:
            score+=1
        if x2!=0:
            score+=1
        if x3!=0:
            score+=1
        if x4!=0:
            score+=1
    score=score/(4.0*mat_data_len) # score相当于非零数据所占比例
    if score>=0.75: # 得分大于设定阈值的视为正常数据    
        normal_mat_info.append((i,score))
    if i%10==0:
        print(f'{i}th round completed')

```

```python
#此处输出格式：(mat文件序号，mat文件总长，正常数据开始时间，正常数据结束时间，正常度得分)
for i in normal_mat_info:
    print(i)
```

```python
normal_mat_indices=[]
for i in normal_mat_info:
    normal_mat_indices.append(i[0])
print(normal_mat_indices)
```
